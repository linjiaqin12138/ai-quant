```python
from typing import List
from datetime import datetime, timedelta
from lib.model import Ohlcv
from lib.history import OhlcvHistory

import talib
import mplfinance as mpf
import numpy as np

import pandas as pd
```

```python
history = OhlcvHistory('BTC/USDT', '1d')
```

```python
last_300_day_data = history.range_query(datetime.now() - timedelta(days=300), datetime.now())
```

```python
class OhlcvHelper:
    def __init__(self, ohlcv_list: List[Ohlcv]):
        df = pd.DataFrame(ohlcv_list)
        df.sort_values(by="timestamp", ascending=True, inplace=True)
        self.df = df

    def get_df(self):
        return self.df

    def macd_info(self):
        _ ,_ ,macd_hist = talib.MACD(df['close'])
        self.df['macdHist'] = macd_hist
        gold_cross_idxs = []
        dead_cross_idxs = []
        bad_turn_good_idxs = []
        good_turn_bad_idxs = []
        for i in range(1, len(self.df)):
            if macd_hist[i] > 0 and macd_hist[i-1] <= 0:
                gold_cross_idxs.append(i)
            elif macd_hist[i] < 0 and macd_hist[i-1] >= 0:
                dead_cross_idxs.append(i)
            elif macd_hist[i] - macd_hist[i-1] > 0 and macd_his[i] < 0:
                bad_turn_good_idxs.append(i)
            elif macd_hist[i] - macd_hist[i-1] < 0 and macd_his[i] > 0:
                good_turn_bad_idxs.append(i)
        return {
            "good_cross_idx": gold_cross_idxs,
            "dead_cross_idxs": dead_cross_idxs,
            "bad_turn_good_idxs": bad_turn_good_idxs,
            "good_turn_bad_idxs": good_turn_bad_idxs
        }

        
```

```python
helper = OhlcvHelper(last_300_day_data)
```

```python
gold_cross, dead_cross = helper.macd_info()
```

```python
mpf.plot(df.set_index('timestamp'), type='candle', style='yahoo',
         title=' SAR Chart', ylabel='Price', ylabel_lower='MACD',
         addplot=[
             mpf.make_addplot(gold_cross, type='scatter', markersize=100, color='y'),
             mpf.make_addplot(dead_cross, type='scatter', markersize=100, color='black'),
             mpf.make_addplot(df['macdHist'], panel=1, color='b', secondary_y=True),
             mpf.make_addplot(df['macdHist'].where(df['macdHist'] >= 0 ), type='bar', panel=1, color='g', alpha=1, secondary_y=True),
             mpf.make_addplot(df['macdHist'].where(df['macdHist'] < 0 ), type='bar', panel=1, color='r', alpha=1, secondary_y=True),
         ],
         figscale=1.5, figsize=(16, 10))
```
