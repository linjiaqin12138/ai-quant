```python
# 逐行打印reason列的内容，如果那一行的buy_point或sell_point不为空, 则追加打印“买”或“卖”
for index, row in df.iterrows():
    if pd.notna(row['buy_point']) or pd.notna(row['sell_point']):
        if pd.notna(row['buy_point']):
            print("买")
        if pd.notna(row['sell_point']):
            print("卖")
        print(f"理由: {row['reason']}")
        
    else:
        print(f"理由: {row['reason']}")
```
