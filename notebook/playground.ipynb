```python
import ccxt

# 创建一个Binance合约交易所实例
binance = ccxt.binance({
    'options': {
        'defaultType': 'future',  # 设置为期货交易
    },
})

# 定义要查询的合约符号
symbol = 'BTC/USDT:USDT'  # 例如：BTC/USDT

# 获取资金费率历史记录
funding_rate_history = binance.fetch_funding_rate_history(
    symbol=symbol,
    since=None,  # 可选，开始时间戳
    limit=None   # 可选，返回记录的数量
)

# 打印资金费率历史记录
for entry in funding_rate_history:
    print(entry)
```
